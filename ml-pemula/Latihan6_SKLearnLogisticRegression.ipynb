**Latihan SKLearn Logistic Regression**

Pada latihan ini kita akan menggunakan logistic regression untuk memprediksi apakah seseorang akan membeli setelah melihat iklan sebuah produk.

**Tahapan Latihan**

1. Ubah dataset menjadi Dataframe.
2. Hapus kolom 'User ID'.
3. Pisahkan atribut dan label.
4. Latih model Logistic Regression.
5. Evaluasi akurasi model.


Unggah dataset Social_Network_Ads.csv pada session storage Google Colab.[Tautan](https://www.kaggle.com/dragonheir/logistic-regression)

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Membaca dataset dan mengubahnya menjadi dataframe
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
# Melihat 5 baris pertama dari dataset.
df.head()
```

```python
# Melihat informasi dataset apakah ada nilai yang kosong pada setiap atribut
df.info()
```

```python
# Hapus kolom yang tidak diperlukan
data = df.drop(columns=['User ID'])

# Jalankan proses one-hot encoding dengan pd.get_dummies()
data = pd.get_dummies(data)
data
```

```python
# Pisahkan atribut dan label
predictions = ['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']
X = data[predictions]
y = data['Purchased']

# Lakukan normalisasi terhadap data yang kita miliki
scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns= X.columns)
scaled_data.head()
```

```python
# Bagi data menjadi train dan test untuk setiap atribut dan label
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)

# Latih model dengan fungsi fit
model = LogisticRegression()
model.fit(X_train, y_train)

# Uji akurasi model
model.score(X_test, y_test)
```
